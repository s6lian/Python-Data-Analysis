<p align="left"> In this homework, I used python pandas, numpy, scipy, sklearn.preprocessing for data normalization.</p>
<p align="left"> As shown in df1, the first 10 rows are selected. </p>

```python
import pandas as pd
import numpy as np
import scipy
from sklearn import preprocessing
df = pd.io.parsers.read_csv(
    'winequality-red.csv',sep =';')
df1= df.loc[0:9,]
df1
```

<p> The table below shows the values with min-max normalized values. As shown, values are all in [0,1]. </p>

```python
minmax_scale = preprocessing.MinMaxScaler().fit(df1)
df_minmax = minmax_scale.transform(df1)
rs2 = pd.DataFrame.from_dict(data=df_minmax)
rs2 = rs2.rename_axis('min-max normalized values')
rs2

```

<p> Then, z-score normalization is implemented. </p>

```python
std_scale =preprocessing.StandardScaler().fit(df1)
df_std = std_scale.transform(df1)
rs1 = pd.DataFrame.from_dict(data=df_std)
rs1 = rs1.rename_axis('z-score normalized values')
rs1
```

<p> Then, I substract each value in the cell with the mean of its column for normalization. </p>

```python
tmp = df1.mean()
rs3 = df1.sub(tmp,axis='columns')
rs3 = rs3.rename_axis('mean subtracted normalized values')
rs3
```

<p>In the following section, I display the max and min distance with manhatten distance,euclidean distance
and cosine distance in three tables below.</p>

```python
cal_array = np.zeros([10,9])
for i in range(0,10):
    for j in range(1,10):
        d = scipy.spatial.distance.cityblock(df1.iloc[i], df1.iloc[j], w=None)
        cal_array[i][j-1]=d

mh_max = np.nanmax(cal_array,axis=1,out=mh_max)
cal_array1 = cal_array[cal_array != 0].reshape((10,8))
mh_min = np.nanmin(cal_array1,axis=1,out=mh_min)
mh_rs = pd.DataFrame(data=[mh_max, mh_min]).T
mh_rs.set_axis(['max_distance','min_distance'],axis = 1, inplace = True)
mh_rs.rename_axis('manhatten distance',axis=0,inplace = True)
mh_rs
```

```python
cal_array2 = np.zeros([10,9])
for i in range(0,10):
    for j in range(1,10):
        d = scipy.spatial.distance.euclidean(df1.iloc[i], df1.iloc[j])
        cal_array2[i][j-1]=d

ed_max = np.nanmax(cal_array2,axis=1,out=ed_max)
cal_array3 = cal_array2[cal_array2 != 0].reshape((10,8))
ed_min = np.nanmin(cal_array3,axis=1,out=ed_min)
ed_rs = pd.DataFrame(data=[ed_max, ed_min]).T
ed_rs.set_axis(['max_distance','min_distance'],axis = 1, inplace = True)
ed_rs.rename_axis('euclidean distance',axis=0,inplace = True)
ed_rs
```

```python
cal_array4 = np.zeros([10,9])
for i in range(0,10):
    for j in range(1,10):
        d = scipy.spatial.distance.cosine(df1.iloc[i], df1.iloc[j])
        cal_array4[i][j-1]=d

cos_max = np.nanmax(cal_array4,axis=1,out=cos_max)
cal_array5 = cal_array4[cal_array4 != 0.0].reshape((10,8))
cos_min = np.nanmin(cal_array5,axis=1,out=cos_min)
cos_rs = pd.DataFrame(data=[cos_max, cos_min]).T
cos_rs.set_axis(['max_distance','min_distance'],axis = 1, inplace = True)
cos_rs.rename_axis('cosine distance',axis=0,inplace = True)
cos_rs
```
