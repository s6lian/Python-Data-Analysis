```python
import pandas
from scipy import stats
df1 = pandas.read_csv('wdbc.csv',header= None)
```

```python
df1 = df1.drop(columns = 1)

rs1 = df1.mean()
rs1 = round(rs1,2)

rs_bcw = pandas.DataFrame(rs1)
rs_bcw.set_axis(['mean'],axis = 1, inplace = True)

arr = []

rs = stats.mode(df1)

for item in rs[0][0]:
    arr.append(round(item,2))

rs_bcw['mode'] = arr

rs3 = df1.skew()
rs3 = round(rs3,2)
rs_bcw['skew'] = rs3

rs4 = df1.std()
rs4 = round(rs4,2)
rs_bcw['std'] = rs4

rs5 = df1.var()
rs5 = round(rs5,2)
rs_bcw['var'] = rs5

rs_bcw.to_csv('rs_wdbc.csv')
```

```python
import pandas
df = pandas.read_csv('wdbc.csv',header= None)
import matplotlib.pyplot as plt
df1 = df.groupby([1]).get_group('M')
df2 = df.groupby([1]).get_group('B')
# df1.hist(column = 6)
plt.hist(df1[6],color='b', histtype='stepfilled', normed=True, label="Malignant")
plt.hist(df2[6],color= 'g', histtype='stepfilled', normed=True, label="Benign", alpha = 0.5)
plt.title("Cell Nucleus's Smoothness in Benign/Malignant")
plt.xlabel("Smoothness")
plt.ylabel("")
plt.legend()
plt.savefig('histograms for smoothness1.png')
```
